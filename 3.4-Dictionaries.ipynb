```python
### iterating over dictionaries

for keys in student.keys():
    print(keys)
```

```python
for values in student.values():
    print(values)
```

```python
for key, value in student.items():
    print(f"{key}:{values}")
```

```python
## nested dictionaries

students={
    "students1":{"name":"Krish","age":32},
    "students2":{"name":"Peter","age":35}
}
print(students)
```

```python
print(students["students2"]["name"])
```

```python
students.item()
```

```python
student.items()
```

```python
for student_id, student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key, value in student_info.items():
        print(f"{key}:{value}")
```

```python
## dictionary comprehension

squares={x:x**2 for x in range(5)}
print(squares)
```

```python
evens = {x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## Practical examples

## Use a dictionary to count the frequency of elements in lists

numbers={1,2,2,3,3,4,4,5,6}
frequency_off_number={}
for number in numbers:
    if number in frequency_off_number:
        frequency_off_number[number]+=1
    else:
        frequency_off_number[number]=1

print(frequency_off_number)
```

```python
## Merging two dictionaries to one dictionaries

dick1 = {'a':1,"b":2}
dick2 = {'c':3,'d':4}
merged_dick={**dick1,**dick2}

print(merged_dick)
```
