```python
## Read a text file and count the number of lines words and characters

def count_in_text(file_path):
  with open(file_path, 'r') as file:
    lines = file.readline()
    line_count = len(lines)
    word_count = sum(len(line.split()) for line in lines)
    character_count = sum(len(line) for line in lines)
    return line_count, word_count, character_count
  
file_path = 'example.txt'

lines, words, characters = count_in_text(file_path)

print(f'Lines: {lines}, Words: {words}, Characters: {characters}')
```

```python
## Checking if a path is a file or a directory

import os
path = '/example.txt'
if os.path.isfile(path):
  print(f"The path '{path}' exists in the directory as a file")
elif os.path.isdir(path):
  print(f"The path '{path}' is a directory")
else:
  print(f"The path is neither a path nor a directory")
```
