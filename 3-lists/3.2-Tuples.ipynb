```python
nested_tuple = ((1,1,2),("a","b","c"),(True, False))

## accessing the elements inside a tuple

print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## printing the nested tuples

for sub_tuple in nested_tuple:
    for values_in in sub_tuple:
        print(values_in, end=" ")
    print()
```
