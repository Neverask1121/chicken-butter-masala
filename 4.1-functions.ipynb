```python
### Variable length arguments
## Positional and keyword arguments

def print_numbers(*krish): ##The star keyword takes up multiple parameters
  for number in krish:
    print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,"Aditya")
```

```python
def print_numbers(*args):
  for number in args:
    print(number)

print_numbers(1,2,3,4,5,6,7)
```

```python
## Keyword arguments
## in keyord arguments all the numbers will be in the key value pair only

def print_numebers_two(**kwargs):
  for key,value in kwargs.items():
    print(f"{key}:{value}")
```

```python
print_numebers_two(name="Aditya", number="67")
```

```python
def print_details(*args, **kwargs):
  for val in args:
    print(f"Positional argument: {val}")

  for key, value in kwargs.items():
    print(f"{key}:{value}") 
```

```python
print_details(1,2,3,4,5,"Aditya", name="Aditya", age="18", country="India")
```

```python
def convert_temperature(temp, unit):
  """This function converts temperature between Celcius and Fahrenheit"""
  if unit == 'C':
    return temp * 9/5 + 32 ## Celcius to Fahrenheit
  elif unit=='F':
    return (temp-32)*5/9 ## Fahrenheit to Celcius
  else:
    return None
  
print(convert_temperature(98, 'F'))
print(convert_temperature(36, 'C'))
  
```

```python
def is_strong_password(password):
  """This function checks whether the password is strong or not"""
  if len(password)<8:
    return False
  if not any(char.isdigit() for char in password):
    return False
  if not any(char.islower() for char in password):
    return False
  if not any(char in '!@#$%^&*()_+' for char in password):
    return False
  return True

## calling function

print(is_strong_password('Weak'))
print(is_strong_password('Strong123@'))
```

```python
## Calculate the total cost of items
## Cart can be given as dictionary or item
def calculate_total_cost(cart):
  total_cost = 0
  for item in cart:
    total_cost+=item['price']*item['quantity']

  return total_cost

## Example cart data

cart=[
  {'name': 'Apple', 'price':0.5, 'quantity':4},
  {'name': 'Banana', 'price':1.5, 'quantity':5},
  {'name': 'Mango', 'price':2.5, 'quantity':6}
]

## calling the function 
total_cost = calculate_total_cost(cart)
print(total_cost)
```

```python
## Checking if a string is Palindrome or not

def is_palindrome(s):
  s=s.lower().replace(" ","")
  return s==s[::-1]

print(is_palindrome("A man a plan a canal Panama"))
print(is_palindrome("Hello"))
```

```python
##Calculate the factorials of a number using recursion

def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)
  
print(factorial(6))
```
