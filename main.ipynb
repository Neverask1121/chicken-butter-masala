```python
num1 = int(input("Enter num1"))
op = input("Enter operation")
num2 = int(input("Enter num2"))

if op == "+":
    result = num1+num2
elif op == "-":
    result = num1-num2
elif op == "/":
    result = num1/num2
elif op == "*":
    result = num1*num2
elif op == "%":
    result = num1%num2

print(result)
```
