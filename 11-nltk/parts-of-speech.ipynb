## Parts of speech tagging

```python
## Speech of Dr. APJ Abdul Kalam

paragraph = """
  Good morning everyone,

It is a privilege to stand before you today. I would like to share a simple thought that has helped me throughout my journey: success is not achieved through talent alone; it is achieved through consistent effort, learning, and resilience.

We live in a world that changes rapidly. New technologies emerge every day, industries evolve, and opportunities appear where none existed before. In such a world, our greatest strength is not what we already know, but our willingness to keep learning. Every challenge we face teaches us something valuable. Every mistake becomes a lesson, and every setback becomes an opportunity to grow stronger.

As students and future professionals, we often focus on big goals—getting into a good college, securing a great job, or building something meaningful. While these goals are important, it is the small actions we take every day that truly determine our future. Reading one more page, solving one more problem, writing one more line of code, or helping one more person may seem insignificant at the moment, but over time these efforts create extraordinary results.

I also believe that success is not just about personal achievement. It is about using our skills and knowledge to create a positive impact on others. Whether through innovation, leadership, or service, we all have the ability to contribute to society in meaningful ways.

As we move forward, let us stay curious, work hard, and never be afraid to take on new challenges. The future belongs to those who are willing to learn, adapt, and persevere.

Thank you, and have a wonderful day.
"""
```

```python
import nltk
sentences = nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
## Apply Stopwords and Filter And then Apply Stemming
## We will find out the Pos tag
from nltk.corpus import stopwords


for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [word for word in words if word not in set(stopwords.words('english'))]
  # sentences[i] = ' '.join(words) converting all text to sentences
  pos_tag = nltk.pos_tag(words)
  print(pos_tag)
```
