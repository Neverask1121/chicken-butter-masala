```python
corpus = """
  Hello Nigga how are you I am fine thank you do you want to suck please suck nigga what the fuck instagram added some new random feature and I want to join google as intern
  Printing the corpus after this
"""
```

```python
print(corpus)
```

```python
from nltk.tokenize import sent_tokenize
sent_tokenize(corpus)

## This only convert sentences to different tokens
```

```python
documents = sent_tokenize(corpus)
type(documents)
# for sentences in documents:
#   print(sentences)
```

```python
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
for sentences in documents:
  print(word_tokenize(sentences))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
