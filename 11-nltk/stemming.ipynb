# Stemming

```python
## Classification problem

words = ["nigga", "eating", "feeding", "sucking", "sleeping", "drinking", "darling", "licking", "chewing", "gooning", "watching", "playing", "chilling"]
```

#### Porter Stemmer


```python
from nltk.stem import PorterStemmer
stemming = PorterStemmer()

for word in words:
  print(word+"---->"+stemming.stem(word))
```
